```python
for i in range(5):
    print(f"{i}:", i * "=", end="")
    if i % 2 == 1:
        print("(odd)")
    else:
        print("(even)")
for i in range(5, -1, -1):
    print(f"{i}:", i * "=", end="")
    if i % 2 == 0:
        print("(even)")
    else:
        print("(odd)")  
```
